```python
# Helper to try normal read first, then fallback if file is malformed

def load_possible_malformed_csv(path):
    """Try reading CSV normally. If the file loads as one-column (everything in a single column),
    attempt to split into columns using the known header layout.
    Returns a DataFrame and a boolean `fixed` which indicates whether we applied fallback parsing."""
    expected_cols = [
        'Row ID','Order ID','Order Date','Ship Date','Ship Mode','Customer ID','Customer Name',
        'Segment','Country','City','State','Postal Code','Region','Product ID','Category','Sub-Category',
        'Product Name','Sales'
    ]
    # Try normal read
    try:
        df = pd.read_csv(path)
        # If we already have expected columns, great
        if set(expected_cols).issubset(df.columns):
            return df, False
        # If the file has a single column that looks like full row strings, fallback
        if df.shape[1] == 1:
            raw = df.iloc[:,0].astype(str)
            # Some files include header in first row; check
            header_row = raw.iloc[0]
            # If header appears to be comma separated of our expected columns, skip it
            # We'll split every row by comma but allow quoted strings
            # Use python's csv reader for robust splitting
            import csv
            rows = []
            for r in raw:
                # csv.reader expects an iterable
                parsed = next(csv.reader([r]))
                rows.append(parsed)
            # Build DataFrame
            max_cols = max(len(r) for r in rows)
            df2 = pd.DataFrame(rows)
            # If header row matches expected columns, set header
            if all(x in df2.iloc[0].tolist() for x in ['Row ID','Order ID','Order Date']):
                df2.columns = df2.iloc[0]
                df2 = df2.drop(df2.index[0]).reset_index(drop=True)
            else:
                # Otherwise assign expected column names if lengths match
                if df2.shape[1] == len(expected_cols):
                    df2.columns = expected_cols
                else:
                    # attempt to trim/merge trailing columns into product name if there are extra columns
                    # e.g. Product Name may contain commas that were not quoted
                    # We'll assume last column(s) belong to Product Name and Sales
                    # Strategy: take first 16 columns as fixed, join any middle columns until last one becomes numeric (sales)
                    first_n = 16  # up to Sub-Category index (0-based 15)
                    tmp = []
                    for r in rows:
                        if len(r) <= first_n+1:
                            # already compatible
                            tmp.append(r + [None]*(first_n+2-len(r)))
                            continue
                        # find last item that can be parsed as float -> that's Sales
                        sales_idx = None
                        for i in range(len(r)-1, first_n-1, -1):
                            try:
                                float(r[i].replace(',',''))
                                sales_idx = i
                                break
                            except:
                                continue
                        if sales_idx is None:
                            # fallback: treat last col as sales
                            sales_idx = len(r)-1
                        # Now build new row: first first_n columns + joined product name (from first_n to sales_idx-1) + sales
                        new_row = r[:first_n] + [','.join(r[first_n:sales_idx])] + [r[sales_idx]]
                        # Ensure length
                        if len(new_row) < len(expected_cols):
                            new_row += [None] * (len(expected_cols)-len(new_row))
                        tmp.append(new_row)
                    df2 = pd.DataFrame(tmp, columns=expected_cols)
            return df2, True
        else:
            # we have multiple columns but not the expected ones; try to coerce
            return df, False
    except Exception as e:
        print('Standard read_csv failed:', e)
        raise


df, fixed = load_possible_malformed_csv(raw_path)
print('Loaded. Fallback parsing used?:', fixed)
print('Initial shape:', df.shape)
```

```python
# ---------------------------
# 2) Initial exploration
# ---------------------------
pd.set_option('display.max_columns', None)
print('\n--- Head (5 rows) ---')
print(df.head())
print('\n--- Info ---')
print(df.info())
print('\n--- Describe (numeric) ---')
print(df.describe(include=[np.number]))
print('\n--- Describe (object) ---')
print(df.describe(include=[object]))
```

```python
# ---------------------------
# 3) Cleaning and type conversions
# ---------------------------
# Strip whitespace from column names
df.columns = [str(c).strip() for c in df.columns]

# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=False)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce', dayfirst=False)

# Convert Sales to numeric
df['Sales'] = df['Sales'].astype(str).str.replace('\\$','',regex=True).str.replace(',','')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# Postal Code to string
df['Postal Code'] = df['Postal Code'].astype(str)

print("‚úÖ Data cleaning completed successfully!")

# Check missing values
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (df.isna().mean()*100).sort_values(ascending=False)
print('\nMissing values (count):')
print(missing.head(20))
print('\nMissing values (%):')
print(missing_pct.head(20))

# Duplicates
dup_count = df.duplicated().sum()
print(f"\nDuplicate rows count: {dup_count}")

# If Order Date missing, drop or flag
if 'Order Date' in df.columns and df['Order Date'].isna().any():
    print('Rows with invalid Order Date:', df[df['Order Date'].isna()].shape[0])
```

```python
# ---------------------------
# 4) Enrich / Feature engineering for time-based analysis
# ---------------------------
# Create time-based features
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month
df['month_name'] = df['Order Date'].dt.strftime('%b')
df['year_month'] = df['Order Date'].dt.to_period('M').dt.to_timestamp()
df['day_of_week'] = df['Order Date'].dt.day_name()
df['quarter'] = df['Order Date'].dt.quarter

# Convert Order ID to string
df['Order ID'] = df['Order ID'].astype(str)

# Average order value (AOV) = total Sales / number of unique orders
order_sales = df.groupby('Order ID', dropna=False)['Sales'].sum()
aov = order_sales.mean()
print(f"\nAverage Order Value (AOV): {aov:,.2f}")

# Refunds: detect negative Sales
refunds = df[df['Sales'] < 0]
refunds_count = len(refunds)
refund_amount = refunds['Sales'].sum()
print(f"Refund rows count (Sales<0): {refunds_count}; refund amount: {refund_amount:,.2f}")

print("‚úÖ Feature engineering completed successfully!")
```

```python
# ---------------------------
# 5) High-level trends: total sales by year and by month
# ---------------------------
# Calculate sales by year and month
sales_by_year = df.groupby('year', as_index=True)['Sales'].sum().sort_index()
sales_by_month = df.groupby('year_month')['Sales'].sum().sort_index()

print('\nSales by year:')
print(sales_by_year)

# Plot yearly totals
plt.figure(figsize=(8,4))
sales_by_year.plot(kind='bar')
plt.title('Total Sales by Year')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# Plot monthly time series
plt.figure(figsize=(12,4))
sales_by_month.plot()
plt.title('Monthly Sales (time series)')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.tight_layout()
plt.show()

print("‚úÖ High-level trends analysis completed!")
```

```python

# ---------------------------
# 6) Growth rates
# ---------------------------
# Year-over-year growth in total sales
yoy = sales_by_year.pct_change().dropna()
print('\nYear-over-Year growth rates:')
print((yoy*100).round(2))

# Monthly growth rates (month over same month previous year and month-over-month)
mom = sales_by_month.pct_change()
yoy_month = sales_by_month.pct_change(12)
print('\nSample month-over-month growth (last 12 months):')
print((mom.tail(12)*100).round(2))
print('\nSample year-over-year by month (last 12 months):')
print((yoy_month.tail(12)*100).round(2))

# Plot monthly growth (MoM)
plt.figure(figsize=(12,4))
(mom*100).plot()
plt.title('Month-over-Month growth (%)')
plt.ylabel('% growth')
plt.axhline(0, color='k', linestyle='--')
plt.tight_layout()
plt.show()

print("‚úÖ Growth rates analysis completed!")
```

```python
# ---------------------------
# 7) Loyalty program performance
# ---------------------------
# Check for loyalty columns
loyalty_cols = [c for c in df.columns if 'loyal' in c.lower() or 'member' in c.lower() or 'loyalty' in c.lower()]
print('\nLoyalty columns found:', loyalty_cols)

# Analyze customer loyalty by repeated purchases
cust_orders = df.groupby('Customer ID')['Order ID'].nunique().sort_values(ascending=False)
top_repeat = cust_orders.head(10)
print('\nTop customers by number of distinct orders:')
print(top_repeat)

# Mark customers with > threshold orders as "likely loyal"
thr = 3
likely_loyal = cust_orders[cust_orders >= thr].index
df['is_likely_loyal'] = df['Customer ID'].isin(likely_loyal)

# Compare sales between loyal and non-loyal customers
loyalty_perf = df.groupby('is_likely_loyal')['Sales'].agg(['sum','count','mean']).sort_values('sum',ascending=False)
print('\nLoyalty performance (inferred):')
print(loyalty_perf)

print("‚úÖ Loyalty analysis completed!")

```

```python
# ---------------------------
# 8) Refund rate and AOV summary
# ---------------------------
if 'Sales' in df.columns:
    total_sales = df['Sales'].sum()
    total_orders = df['Order ID'].nunique() if 'Order ID' in df.columns else len(df)
    refunds_amount = refund_amount
    refunds_rate = abs(refunds_amount) / total_sales if total_sales and refunds_amount<0 else 0
    print(f"\nTotal Sales: {total_sales:,.2f}")
    print(f"Total Orders: {total_orders:,}")
    print(f"Refund amount: {refunds_amount:,.2f}")
    print(f"Refund rate (refund amt / total sales): {refunds_rate:.2%}")
    if aov is not None:
        print(f"AOV: {aov:,.2f}")
```

```python
# ---------------------------
# 9) Dimensional segmentation
# ---------------------------
# Top regions, states, categories, sub-categories, products
dimensions = ['Region','State','Category','Sub-Category','City']

for d in dimensions:
    print(f"\nTop 10 {d} by Sales:")
    top_10 = df.groupby(d)['Sales'].sum().sort_values(ascending=False).head(10)
    print(top_10)
    plt.figure(figsize=(8,4))
    top_10.plot(kind='bar')
    plt.title(f'Top 10 {d} by Sales')
    plt.ylabel('Sales')
    plt.tight_layout()
    plt.show()

# Concentration: what share of sales come from top X% of products or customers
prod_sales = df.groupby('Product ID')['Sales'].sum().sort_values(ascending=False)
cum = prod_sales.cumsum() / prod_sales.sum()
top_10pct_cut = prod_sales[cum <= 0.10].sum()
print(f"\nSales from top products that make up 10% cumulative: {top_10pct_cut:,.2f}")

cust_sales = df.groupby('Customer ID')['Sales'].sum().sort_values(ascending=False)
top_customers_share = cust_sales.head(int(max(1, len(cust_sales)*0.1))).sum() / cust_sales.sum()
print(f"Share of sales from top 10% customers (approx): {top_customers_share:.2%}")

print("‚úÖ Dimensional segmentation completed!")
```
